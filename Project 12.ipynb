```python
import pandas as pd
from sklearn.cluster import KMeans

# -----------------------------
# Step 1: Create Sample Customer Data
# -----------------------------
data = {
    "CustomerID": [101,102,103,104,105,106,107,108,109,110],
    "Age": [22,25,35,28,40,31,27,45,38,29],
    "AnnualIncome": [25000,30000,50000,35000,80000,45000,40000,90000,70000,38000],
    "SpendingScore": [80,75,45,70,20,50,65,15,30,60]
}

df = pd.DataFrame(data)

print("Original Dataset")
print(df)

# -----------------------------
# Step 2: Customer Segmentation
# -----------------------------
X = df[["AnnualIncome", "SpendingScore"]]

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Segment"] = kmeans.fit_predict(X)

print("\nCustomer Segments")
print(df)

# -----------------------------
# Step 3: Customer Behavior Analysis
# -----------------------------
print("\nSegment Summary")
print(df.groupby("Segment")[["AnnualIncome", "SpendingScore"]].mean())

print("\nNumber of Customers in Each Segment")
print(df["Segment"].value_counts())

# -----------------------------
# Step 4: Targeting Strategy
# -----------------------------
print("\nTargeting Strategy")
print("Segment 0 : Premium customers - Offer exclusive rewards.")
print("Segment 1 : Regular customers - Provide loyalty discounts.")
print("Segment 2 : High spending customers - Recommend premium products.")

# -----------------------------
# Step 5: Business Recommendations
# -----------------------------
print("\nBusiness Recommendations")
print("1. Offer personalized marketing campaigns.")
print("2. Reward loyal customers with special benefits.")
print("3. Promote premium products to high-value customers.")
print("4. Provide discounts to improve customer engagement.")

print("\nCapstone Project 2 Completed Successfully!")
```
